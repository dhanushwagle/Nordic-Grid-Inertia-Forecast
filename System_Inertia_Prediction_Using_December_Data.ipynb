```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
act = pd.read_csv('December-2019.csv')
pred = pd.read_csv('PredictOutput.csv')
```

```python
act.tail()
```

```python
pred.tail()
```

```python
# KE plot 
plt.figure()
x1 = range(0,44640)
plt.plot(x1,act['KE'],color='red',label='Actual')
x2 = range(0,44670)
plt.plot(x2,pred['yhat'],color='blue',label='Predicted')
plt.legend(loc='best', fontsize='x-large')
plt.xlabel('Time(m)')
plt.ylabel('GWs')
plt.title('Actual vs Predicted KE values')
plt.show()
```

```python
#inertia plot

S_sys = 52.013812843 # GW
t = 1
H_act = []
H_pred = []

for values in act['KE']:
    H_system = values/S_sys
    H_act.append(H_system)
    t+=1


for values in pred['yhat']:
    H_system = values/S_sys
    H_pred.append(H_system)
    t+=1

plt.figure()
x1 = list(range(0, 44640))
plt.plot(x1,H_act,'r-', label='Actual')
x2 = list(range(0, 44670))
plt.plot(x2,H_pred,'g-', label='Predicted')
plt.legend(loc='best', fontsize='x-large')
plt.title('Actual vs Predicted Sytem Inertia')
plt.xlabel('Time(min)')
plt.ylabel('Inertia (sec)')
plt.savefig('Inertia_comparision.jpeg', dpi=300)
plt.show()
```
